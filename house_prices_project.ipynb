```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('house_prices_realistic.xlsx')
```

```python
print(df.head(10))
```

```python
print(df.tail(10))
```

```python
print(df.shape)
print(df.columns)
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
print(df.describe())
```

```python
print(df.select_dtypes(include=np.number).agg(['min', 'max']))
```

```python
df.groupby('Year')['Price'].mean().plot(kind='bar')
plt.title("Average House Prices Per Year")
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```

```python
df.groupby('Year')['Price'].mean().plot(kind='line', marker='o')
plt.title("Price Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
df['Price'].hist(bins=20)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['Size_m2'], df['Price'])
plt.title("Size vs Price")
plt.xlabel("Size (m2)")
plt.ylabel("Price")
plt.show()
```

```python
df.groupby('Bedrooms')['Price'].mean().plot(kind='bar')
plt.title("Bedrooms vs Price")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()
```

```python
df.groupby('City')['Price'].mean().plot(kind='bar')
plt.title("City vs Price")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.show()
```

```python
print("""
INSIGHTS:
- Prices increase as house size increases.
- More bedrooms generally lead to higher prices.
- Some cities have higher average prices than others.
- House prices change over different years.
- There is a clear relationship between price and property features.
""")
```
